```python
import numpy as np

def kmeans(dataset, k):
    """
    Performs k-means clustering on the given dataset.
    """
    np.random.seed(42)
    # 1. Initialize centers randomly from the data points for better results
    random_indices = np.random.choice(dataset.shape[0], size=k, replace=False)
    centers = dataset[random_indices]
    
    clusters_old = np.zeros(dataset.shape[0])
    
    # Loop until the assignments stop changing
    while True:
        # 2. Assign each data point to the closest center (Assignment Step)
        distances = np.linalg.norm(dataset[:, np.newaxis] - centers, axis=2)
        clusters = np.argmin(distances, axis=1)
        
        # 3. Check for convergence
        if np.array_equal(clusters, clusters_old):
            break
            
        clusters_old = clusters.copy()
        
        # 4. Update centers to be the mean of their assigned points (Update Step)
        new_centers = np.zeros((k, dataset.shape[1]))
        for c_idx in range(k):
            # Get all points assigned to this cluster
            points_in_cluster = dataset[clusters == c_idx]
            
            # If a cluster is not empty, calculate its new center
            if len(points_in_cluster) > 0:
                new_centers[c_idx] = np.mean(points_in_cluster, axis=0)
        
        centers = new_centers
        
    return clusters, centers
```

```python
def kmeansplusplus_custom(dataset, k):
    """
    1. Initialize the first center randomly from the data points
    2. For each remaining center:
        - Calculate the distance from each data point to the nearest center
        - Choose the next center with probability proportional to the squared distance
    3. Repeat steps 2-3 until k centers are selected
    4. Initialize the clusters
    5. Repeat steps 6-7 until convergence:
        - Assign each data point to the nearest center
        - Update the centers to be the mean of their assigned points
    """
    # Initialize the first center index randomly from the data points
    np.random.seed(42)
    random_index = np.random.choice(dataset.shape[0])
    # Access the data point at the random index as first center
    first_center = dataset[random_index]
    # initialize the list of distances from each data point to the respective center
    distances_list = []
    # Centers list initilize with first center
    centers = [first_center]
    
    
    for _ in range(k-1):
        distances = np.linalg.norm(dataset[:, np.newaxis] - centers[-1], axis=2)
        distances_list.append(distances)
        scores = np.min(distances_list, axis=0)
        scores = scores**2
        probabilities = scores / scores.sum()
        index = np.argmax(probabilities)
        centers.append(dataset[index])
    
    clusters_old = np.zeros(dataset.shape[0])
    
    # Loop until the assignments stop changing
    while True:
        # 2. Assign each data point to the closest center (Assignment Step)
        distances = np.linalg.norm(dataset[:, np.newaxis] - centers, axis=2)
        clusters = np.argmin(distances, axis=1)
        
        # 3. Check for convergence
        if np.array_equal(clusters, clusters_old):
            break
            
        clusters_old = clusters.copy()
        
        # 4. Update centers to be the mean of their assigned points (Update Step)
        new_centers = np.zeros((k, dataset.shape[1]))
        for c_idx in range(k):
            # Get all points assigned to this cluster
            points_in_cluster = dataset[clusters == c_idx]
            
            # If a cluster is not empty, calculate its new center
            if len(points_in_cluster) > 0:
                new_centers[c_idx] = np.mean(points_in_cluster, axis=0)
        
        centers = new_centers

    return centers, clusters

```

```python
def kmeansplusplus(dataset, k):
    np.random.seed(42)
    # 1. Initialize the first center randomly
    centers = [dataset[np.random.choice(dataset.shape[0])]]
    
    # 2. Initialize remaining centers using k-means++ logic
    for _ in range(1, k):
        # Calculate squared distances to nearest center for each point
        distances = np.array([min([np.linalg.norm(x - c) ** 2 for c in centers]) 
                            for x in dataset])
        
        # Convert distances to probabilities
        probabilities = distances / distances.sum()
        
        # Choose next center with probability proportional to distance squared
        next_center_idx = np.random.choice(len(dataset), p=probabilities)
        centers.append(dataset[next_center_idx])
    
    centers = np.array(centers)
    clusters_old = np.zeros(dataset.shape[0])
    
    # Main k-means loop
    while True:
        # Assignment step
        distances = np.linalg.norm(dataset[:, np.newaxis] - centers, axis=2)
        clusters = np.argmin(distances, axis=1)
        
        # Check for convergence
        if np.array_equal(clusters, clusters_old):
            break
            
        clusters_old = clusters.copy()
        
        # Update step
        new_centers = np.zeros((k, dataset.shape[1]))
        for c_idx in range(k):
            points_in_cluster = dataset[clusters == c_idx]
            if len(points_in_cluster) > 0:
                new_centers[c_idx] = np.mean(points_in_cluster, axis=0)
            else:
                # Handle empty clusters by reinitializing randomly
                new_centers[c_idx] = dataset[np.random.choice(dataset.shape[0])]
        
        centers = new_centers

    return centers, clusters
```

# evaluating for finding similarities between our code and sklearn kmeans

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from sklearn.cluster import KMeans
import numpy as np

def evaluate_clustering(X, k, custom_centers, custom_labels, random_state=42):
    """
    Evaluate and compare custom k-means++ implementation with scikit-learn's implementation.
    
    Parameters:
    - X: Input data (numpy array)
    - k: Number of clusters
    - custom_centers: Cluster centers from your implementation
    - custom_labels: Cluster assignments from your implementation
    - random_state: Random seed for reproducibility
    
    Returns:
    - Dictionary containing evaluation metrics for both implementations
    """
    results = {}
    
    # Calculate metrics for custom implementation
    if len(np.unique(custom_labels)) > 1:  # Silhouette score requires more than 1 cluster
        results['custom'] = {
            'inertia': np.sum([np.sum((X[custom_labels == i] - custom_centers[i])**2) 
                             for i in range(k)]),
            'silhouette': silhouette_score(X, custom_labels),
            'calinski_harabasz': calinski_harabasz_score(X, custom_labels),
            'davies_bouldin': davies_bouldin_score(X, custom_labels)
        }
    else:
        results['custom'] = {
            'inertia': np.sum([np.sum((X[custom_labels == i] - custom_centers[i])**2) 
                             for i in range(k)]),
            'silhouette': None,
            'calinski_harabasz': None,
            'davies_bouldin': None
        }
    
    # Run scikit-learn's KMeans for comparison
    kmeans_sklearn = KMeans(n_clusters=k, init='k-means++', n_init=10, 
                           random_state=random_state)
    sklearn_labels = kmeans_sklearn.fit_predict(X)
    
    results['sklearn'] = {
        'inertia': kmeans_sklearn.inertia_,
        'silhouette': silhouette_score(X, sklearn_labels),
        'calinski_harabasz': calinski_harabasz_score(X, sklearn_labels),
        'davies_bouldin': davies_bouldin_score(X, sklearn_labels)
    }
    
    return results

def print_evaluation_results(results):
    """Helper function to print evaluation results in a readable format."""
    print("Clustering Evaluation Results")
    print("=" * 50)
    
    for model, metrics in results.items():
        print(f"\n{model.upper()} Implementation:")
        print("-" * 30)
        for metric, value in metrics.items():
            if value is not None:
                if isinstance(value, float):
                    print(f"{metric.replace('_', ' ').title()}: {value:.4f}")
                else:
                    print(f"{metric.replace('_', ' ').title()}: {value}")
            else:
                print(f"{metric.replace('_', ' ').title()}: Not applicable (only one cluster)")

# Example usage:
# 1. First run your k-means++ implementation
centers, cluster_assignments = kmeansplusplus(X_scaled, k=6)
# 
# 2. Then evaluate both implementations
results = evaluate_clustering(X_scaled, k=6, 
                            custom_centers=centers, 
                            custom_labels=cluster_assignments)
# 
# 3. Print the results
print_evaluation_results(results)
```

```python
# Example usage:
# 1. First run your k-means++ implementation
centers, cluster_assignments = kmeansplusplus_custom(X_scaled, k=6)
# 
# 2. Then evaluate both implementations
results = evaluate_clustering(X_scaled, k=6, 
                            custom_centers=centers, 
                            custom_labels=cluster_assignments)
# 
# 3. Print the results
print_evaluation_results(results)
```
